```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_iris

data = load_iris()
X = data.data
y = data.target
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

print(df.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Akurasi
acc = accuracy_score(y_test, y_pred)
print("\nAkurasi:", acc)
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = model.predict(np.c_[xx.ravel(), yy.ravel(),
                        np.zeros_like(xx.ravel()),
                        np.zeros_like(xx.ravel())])

Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("KNN Decision Boundary (Iris Dataset)")
plt.show()
```

Model KNN digunakan untuk mengklasifikasikan jenis bunga Iris berdasarkan fitur yang ada. Dengan nilai K=3, model menentukan kelas berdasarkan 3 tetangga terdekat. Hasil akurasi menunjukkan bahwa model cukup baik dalam mengklasifikasikan data.

KNN merupakan metode sederhana namun efektif untuk klasifikasi. Dengan dataset yang sesuai, model dapat menghasilkan akurasi yang cukup tinggi.
