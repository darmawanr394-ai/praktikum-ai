```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X = np.array([
    [1,60],
    [2,65],
    [3,70],
    [4,75],
    [5,80],
    [6,85]
])

y = np.array([0,0,0,1,1,1])

model = LogisticRegression()
model.fit(X,y)
```

```python
pred = model.predict([[3,72]])
print("Prediksi (0=tidak lulus, 1=lulus):", pred)
```

```python
X1 = X[:,0].reshape(-1,1)

model_vis = LogisticRegression()
model_vis.fit(X1,y)
```

```python
x_range = np.linspace(1,6,100).reshape(-1,1)
y_prob = model_vis.predict_proba(x_range)[:,1]
```

```python
plt.scatter(X1,y)
plt.plot(x_range,y_prob)
plt.title("Regresi Logistik (Probabilitas Lulus)")
plt.xlabel("Jam Belajar")
plt.ylabel("Probabilitas")
plt.show()
```

Grafik menunjukkan hubungan antara jam belajar dan probabilitas kelulusan. Semakin tinggi jam belajar, peluang lulus semakin besar. Model logistik membentuk kurva sigmoid yang umum pada klasifikasi.
