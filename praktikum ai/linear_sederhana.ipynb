```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1,2,3,4,5,6]).reshape(-1,1)
y = np.array([50,55,60,65,70,75])
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.title("Regresi Linear Sederhana")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai")
plt.show()

print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
```

Dari hasil model terlihat bahwa terdapat hubungan positif antara jam belajar dan nilai. Semakin lama waktu belajar, nilai cenderung meningkat secara linear. Model ini cukup sederhana namun mampu menggambarkan pola data dengan baik.
