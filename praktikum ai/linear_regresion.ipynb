```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.array([
    [1,5],
    [2,6],
    [3,5],
    [4,7],
    [5,6],
    [6,7]
])

y = np.array([50,58,62,70,75,80])

model = LinearRegression()
model.fit(X,y)

# Ambil variabel
x1 = X[:,0]
x2 = X[:,1]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1, x2, y)

ax.set_xlabel("Jam Belajar")
ax.set_ylabel("Jam Tidur")
ax.set_zlabel("Nilai")

plt.title("Regresi Linear Berganda (3D)")
plt.show()

print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
```

Model menunjukkan bahwa nilai dipengaruhi oleh dua faktor yaitu jam belajar dan jam tidur. Jam belajar memiliki pengaruh lebih besar dibandingkan jam tidur terhadap peningkatan nilai. Dengan dua variabel, model menjadi lebih akurat dibandingkan regresi sederhana.
